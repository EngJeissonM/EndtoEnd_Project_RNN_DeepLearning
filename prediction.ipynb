```python
# Step 1: Import Libraries and Load the Model
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb # IMDB dataset for sentiment analysis
from tensorflow.keras.preprocessing import sequence # for padding sequences, this helps in making all input sequences of same length
from tensorflow.keras.models import load_model # to load pre-trained models
```

```python
# Load the IMDB dataset word index
word_index = imdb.get_word_index() # to get the mapping of words to their integer indices
reverse_word_index = {value: key for key, value in word_index.items()} # to reverse the mapping for decoding reviews
```

```python
# Load the pre-trained model with ReLU activation
model = load_model('simple_rnn_imdb.h5') # loading the saved Simple RNN model
model.summary()
```

```python
model.get_weights() # to view the weights of the model
```

```python
# Step 2: Helper Functions
# Function to decode reviews back to text
def decode_review(encoded_review):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])
 # Subtracting 3 to account for reserved indices

# Function to preprocess user input
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
### Prediction  function

def predict_sentiment(review):
    preprocessed_input=preprocess_text(review) #send the review for preprocess_text function

    prediction=model.predict(preprocessed_input) #get the prediction from the model

    sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'
    
    return sentiment, prediction[0][0]

```

```python
# Step 4: User Input and Prediction
# Example review for prediction
example_review = "This movie was fantastic! The acting was great and the plot was thrilling."

sentiment,score=predict_sentiment(example_review)

print(f'Review: {example_review}')
print(f'Sentiment: {sentiment}')
print(f'Prediction Score: {score}')
```
